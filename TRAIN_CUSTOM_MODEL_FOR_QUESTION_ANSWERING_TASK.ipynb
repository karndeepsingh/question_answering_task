# QUESTION ANSWERING TASK 
The goal of Question Answering is to find the answer to a question given a question and an accompanying context. 

# DATA FORMATE REQUIRED FOR TRAINING QUESTION ANSWERING TASK


**Each such dictionary contains two attributes, the "context" and "qas".**

1.**context:** The paragraph or text from which the question is asked.

2.**qas:** A list of questions and answers (format below).


**Questions and answers are represented as dictionaries. Each dictionary in qas has the following format.**

1.id: (string) A unique ID for the question. Should be unique across the entire dataset.

2.question: (string) A question.

3.is_impossible: (bool) Indicates whether the question can be answered correctly from the context.

4.answers: (list) The list of correct answers to the question.

**A single answer is represented by a dictionary with the following attributes.**

**1.text: (string)** The answer to the question. Must be a substring of the context.

**2.answer_start:** (int) Starting index of the answer in the context.

```python
#Loading the data and converting into the required format as shown above
import json


with open('/content/train-v2.0.json', 'r') as f:
    train_data = json.load(f)

train_data = [item for topic in train_data['data'] for item in topic['paragraphs'] ]
```

```python
train_data[0]
```

```python
from simpletransformers.question_answering import QuestionAnsweringModel


train_args = {
    'learning_rate': 3e-5,
    'num_train_epochs': 2,
    'max_seq_length': 384,
    'doc_stride': 128,
    'overwrite_output_dir': True,
    'reprocess_input_data': False,
    'train_batch_size': 6,
    'gradient_accumulation_steps': 8,
}

model = QuestionAnsweringModel('bert', 'bert-base-uncased', args=train_args)
```

```python
model.train_model(train_data[:50])
```

```python
with open('/content/dev-v2.0.json', 'r') as f:
    dev_data = json.load(f)

dev_data = [item for topic in dev_data['data'] for item in topic['paragraphs'] ]

```

```python
preds = model.predict(dev_data[:10])
```

```python
dev_data[0]
```

```python
import os
os.makedirs('results', exist_ok=True)

submission = {i['id']: i['answer'] for pred in preds for i in pred for key,value in i.items() if key == "answer"}

with open('results/submission.json', 'w') as f:
    json.dump(submission, f)
```
